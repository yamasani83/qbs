````python
fig = plt.figure(figsize = (10,6))      # sets overall figure size
ax = plt.axes((0.1,0.1,0.5,0.8))      # sets extents of figure on canvas
ax.spines['right'].set_visible(False)   #removes right spine / border line
ax.spines['left'].set_visible(False)
ax.spines['bottom'].set_visible(False)
ax.spines['top'].set_visible(False)
ax.tick_params(axis = 'x', which = 'major', direction = 'out', length = 10, 
               width = 5,color = 'red', pad = 15, labelsize = 15, labelcolor = 'green',
               labelrotation = 15)         #sets parameters for x axis
ax.set_xticks([0.2,1])                     #set the position of x-axis tick labels
ax.set_xticklabels(['pos1','pos2'])        #change the name of x-axis tick labels
ax.yaxis.set_major_locator(MultipleLocator(0.5))    #sets spacing b/w major y-axis tick lines / labels
ax.yaxis.set_minor_locator(MultipleLocator(0.1))    #set spacing b/w of major y-axis tick lines / labels
````

## Creating a Line Plot

````python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(flights.month_num, flights.passengers)
````

````python
fig = plt.figure(figsize=(10,6))
ax = fig.add_axes([0.1,0.1,0.5,0.8])
ax.plot(flights.month_num, flights.passengers,   # data
marker='o',     # each marker will be rendered as a circle
markersize=4,   # marker size
markerfacecolor='red',   # marker facecolor
markeredgecolor='black',  # marker edgecolor
markeredgewidth=1,       # marker edge width
linestyle='--',            # line style will be dash line
linewidth=1)          # line width
````
