# Visualization 

## Themes, Styles, and Formatting

In the previous lesson, we learned how easy it is to create some basic plots using **seaborn**, which is built on top of the more advanced **matplotlib**. 

When we reach the publication stage of our research, however, we may wish to further customize our visualizations. We can do so in the following ways:
+ modify the parameters of an individual plot
+ modify the default parameters to change all plots within a document
+ change the default theme for all plots displayed within a document

1. Let's begin by importing the necesary packages and reading in the Titanic dataset and creating a swarmplot.




````python
import matplotlib as mpl
import matplotlib.pyplot as plt
from matplotlib.ticker import MultipleLocator,FormatStrFormatter,MaxNLocator
import pandas as pd
import seaborn as sns
from pathlib import Path
import os
````

````python
titanic = sns.load_dataset("titanic")
````

````python
plt.scatter(data = titanic, x = "age", y = "fare", c = "survived")
````

````python
p = plt.scatter(data = titanic, x = "age", y = "fare", s = "fare", c = "survived", alpha = 0.5)
````

````python
plt.rcParams['figure.figsize'] = [25, 10]  #sets size of the visualizations in our notebook
%config InteractiveShellApp.matplotlib = 'inline'
%config InlineBackend.figure_formats = ['svg']
````

````python
p.figure
````

````python
titanic = sns.load_dataset("titanic")
sw = sns.swarmplot(data = titanic, x = "sex", y = "age", hue = "survived")
````

````python
sw.figure
````

## Style / Theme

2. The easiest way to change the general look of your plots it to load a pre-set stylesheet. [Matplotlib has a set of style sheets. Read here for more.](https://www.python-graph-gallery.com/199-matplotlib-style-sheets)

To see a full list of available matplotlib style sheets, run:

````python
plt.style.available
````

````python
mpl.style.use
````

2b. We can apply one of these style sheets by simply adding

```
plt.style.use([name of style sheet])
```

Run the code below, then try replacing the 'fivethirtyeight' stylesheet with others. Which one do you like the best?

````python
plt.style.use('fivethirtyeight')
sw.figure
````

2c. Note, once we change the style sheet, it will apply to all other plots in this document until we change it. For example, run the code below and see how it follows the same style as above.

````python
sns.countplot(data = titanic, x = "survived")
````
